### 欧拉公式

$$e^{i\theta}=\cos\theta + \sin\theta i$$

#### 1. $(1+xi)^n$ 的几何含义
- 复数 x 乘 y 的几何含义为**将 x 辐角增加旋转 y 辐角，再将 x 向量的模缩放为原来的 |y| 倍**
- 从 1 ，即复平面上向量 (1, 0) 开始，乘 $(1+xi)$ 即旋转 $\arctan(x)$，缩放为原来的 $\sqrt{1+x^2}$ 倍，从平面上看就是构建一个直角三角形，原向量为其中一条直角边，模长为 1，另一条直角边模长为 x，乘 $(1+xi)$ 后变成了这个三角形的斜边
- 继续乘 $(1+xi)$ 表示以前一个三角形的斜边作为一条直角边，构建另一个**相似**直角三角形，新三角形的斜边即乘积（为什么？因为每次旋转幅角一样，每次放缩程度也一样）
- 即 $(1+xi)^n$ 表示，以 1 和 x 为直角边构建一个直角三角形，在此基础上依次构建 n 个相似直角三角形，最后一个三角形的斜边表示的向量

#### 2. $e^{i\theta}$ 的几何含义
- 按照定义 $e^{i\theta}$ 表示为 $\lim_{n \rightarrow \infty} (1 + \frac{i\theta}{n})^n$
- 从上面可知 $(1 + \frac{i\theta}{n})^n$ 表示以 1 和 $\frac{\theta}{n}$ 为直角边构建一个直角三角形，在此基础上依次构建 n 个相似直角三角形，最后一个三角形的斜边表示的向量
- 当 n 趋向于无穷大时，$\frac{\theta}{n}$ 趋向于 0，直角三角形的斜边趋向于 1，直角三角形的堆叠趋近于在向量在复平面绕着原点旋转，旋转的圆弧趋近于 $n\times\frac{\theta}{n}=\theta$
- 即 $e^{i\theta}$ 的几何含义为：复平面上模长为 1 的向量从 (1,0) 出发旋转 $\theta$，对应可以写出旋转后的复数为 $\cos\theta+\sin\theta i$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_triangles(theta, n):
  
  def cal(a, i):
    l = (a[0]**2+a[1]**2)**0.5
    lp = l * ((1+(theta/n)**2)**0.5)  # 斜边长度
    t = np.arctan(theta/n)
    return lp*np.cos(i*t), lp*np.sin(i*t)
    
  s = (1, 0)
  plt.figure(figsize=(6, 6))
  plt.xlim(-0.1, 1.1)
  plt.ylim(-0.1, 1.1)
  plt.title(f"n={n}")
  for i in range(n):
    plt.plot((0, s[0]), (0, s[1]), color="grey", alpha=0.2+(i/n)*0.8)
    s2 = cal(s, i+1)
    plt.plot((s[0], s2[0]), (s[1], s2[1]), color="grey", alpha=0.2+(i/n)*0.8)
    s = s2
  plt.plot((0, s[0]), (0, s[1]), color="grey")
  plt.show()
  print(s)
```

```python
theta = np.pi / 3
plot_triangles(theta, n=5)
plot_triangles(theta, n=50)
plot_triangles(theta, n=500)
```
